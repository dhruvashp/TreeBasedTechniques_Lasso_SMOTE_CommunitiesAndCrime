The entire dataset (train + test) will be used here
CV includes only all the features, obviously, not the output class

```python
import pandas as pd
import numpy as np

df = pd.read_csv('aps_df_Final_Imputed_Whole.csv',index_col=0)
print(df)


col_header = df.columns
feature_header = col_header[1:171]
CV = pd.DataFrame(index=np.arange(0,1),columns = feature_header)

i=0
for i in np.arange(1,171):
    std = df.iloc[:,i].std()
    mean = df.iloc[:,i].mean()
    CV.iloc[0,i-1] = std/mean
    

print(CV)
CV.to_csv('aps_features_whole_CV_matrix.csv')
```
