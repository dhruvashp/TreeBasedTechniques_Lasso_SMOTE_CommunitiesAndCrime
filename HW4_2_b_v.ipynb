```python
import pandas as pd
import numpy as np

df = pd.read_csv('aps_df_Final_Imputed_Whole.csv',index_col=0)

total_pos = 0
total_neg = 0
i=0
for i in np.arange(0,76000):
    if df.iloc[i,0] == 0:
        total_neg = total_neg + 1
    else:
        total_pos = total_pos + 1
        

ratio = (total_pos/total_neg)*100

print('The percentage of positive to negative class in the entire dataset is : \n',ratio)

total_pos_train = 0
total_neg_train = 0
i=0
for i in np.arange(0,60000):
    if df.iloc[i,0] == 0:
        total_neg_train = total_neg_train + 1
    else:
        total_pos_train = total_pos_train + 1
        
ratio_train = (total_pos_train/total_neg_train)*100

print('The percentage of positive to negative class in the training set is : \n',ratio_train)

aps_train = df.iloc[0:60000,:]
aps_train.to_csv('aps_train.csv')

aps_test = df.iloc[60000:76000,:]
aps_test.reset_index(drop=True,inplace=True)
aps_test.to_csv('aps_test.csv')

print('The final training set is : \n',aps_train)
print('The final test set is : \n',aps_test)
```

This dataset has an extreme/severe class balance, as can be seen from the percentage of positive to negative class. The class balance here is extremely concerning and will most definitely affect the classification.

The final imputed training and test sets have also been extracted to appropriate files here (.csv) for use in subsequent portions
