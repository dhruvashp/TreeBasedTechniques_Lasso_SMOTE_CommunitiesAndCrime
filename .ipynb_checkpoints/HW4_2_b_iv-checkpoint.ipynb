Scatter Plot Conclusions :

In general as we again have plots of variables with higher CV, taking two features with this high CV, against the classes 0 and 1 again gives a better 'distinguishing power', at least over different classes. Over same classes, this high CV means that despite the class being same, the features still vary comparitively more than other features over the same class. 

Most ideal decision boundary may have features that stay constant throughout a class, thus giving sharp seperation, and extremely crisp decision boundaries, however this usually never happens. If the variability is high, and this variation is in a range that's distinct for distinct classes, then we do have pretty good decision boundaries even then.

This idea can be obtained via boxplots. Thus high variation in features must, for classification, be such that over different classes, this variation is in distinct ranges. We then will have better decision boundaries. Features that don't vary as much over same class, may also not vary from one class to the next thus making the model lose its ability to distinguish between classes. Thus overall low sd/mean, over entire feature set, may imply low variability over classes, thus implying that model based on that feature likely won't classify correctly

High sd/mean may imply overall large variation over different classes, however unless that is supplemented with variation in 'range of variation' over different classes as well, we still may end up having poor model classifying strength/capability. For an idea of this we'll get the boxplots, for a single feature in the two classes.

If sd/mean is extraordinarily high then variation over different classes may be true, but variation over the same class may also be true implying that the classification points, with respect to that feature, are extremely spread and overlapping in nature leading to poor decision boundaries

```python
for i in np.arange(0,13):
    sns.boxplot(data=final_df,y=new_header[i],x='class')
    plt.show()
```

From the boxplots again we run into similar interpretability issues

As such from the specific case boxplot, we see that there does seem to be a legitimate seperation between values for classes 0 and 1, which points to the fact that in this case features with high sd/mean may have good interclass variability, along with an intra-class variability that lies in a range different from that of the other class.
