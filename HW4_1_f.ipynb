```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.read_csv('Final_Imputed_Whole.csv',index_col=0)
print(df)
```

```python
df_train = df.iloc[0:1495,:]
df_test = df.iloc[1495:1994,:]
print(df_train)
print(df_test)
```

```python
df_test.reset_index(drop=True,inplace=True)
print(df_test)
```

```python
X_train = df_train.drop(columns=['ViolentCrimesPerPop'])
X_test = df_test.drop(columns=['ViolentCrimesPerPop'])
y_train = df_train['ViolentCrimesPerPop']
y_test = df_test['ViolentCrimesPerPop']
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
reg = LinearRegression().fit(X_train,y_train)
y_predicted = reg.predict(X_test)
print(y_test)
print(y_predicted)
```

Due to imputation we do get a minor amount of values negative which are extremely close to zero. Negative values, as such, are impossible for our outcome in general.

```python
MSE = mean_squared_error(y_test,y_predicted)
print(MSE)
```

The test MSE thus obtained is 0.0566. As such, while this value may seem rather small, it is still relatively noticeable given that our data has been normalized.

Square root of MSE, gives, on average, deviation (absolute) of the predicted value from the true value, in this case about 0.23, which is quite excessive given that output lies in the range of (0,1) (normalized)
